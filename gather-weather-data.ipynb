```python
import pandas as pd
import numpy as np
import calendar
```

```python
# Download the US weather data from ncdc.noaa.gov
def get_weather_data(state_code, parameter, month):
    """ state_code is an integer value between 1 and 50, 
        parameter is type of weather data i.e average temperature, precipitation etc., 
        month is an integer value between 1 and 12.  
    """
    url = 'https://www.ncdc.noaa.gov/cag/time-series/us/%02d/00/%s/1/%02d/1960-2011.csv?base_prd=true&begbaseyear=1925&endbaseyear=2000' %(state_code, parameter, month)
    print (state_code, parameter, month)
    df = pd.read_csv(url)
    state = df.columns[0]
    print (state)
    df['state'] = [state for i in range(df.shape[0])]
    columns = list(df.iloc[3])[:3] + ['state']
    df.head()
    data = np.asarray(df.iloc[4:,:])
    return pd.DataFrame(data = data, columns=columns)
```

```python
par = ['pcp'] #,'cdd','hdd','pdsi','ph
```

```python
tf = pd.DataFrame()
for i in range(1, 51):
    for j in range(1, 13):
        for k in par:
            try:
                tmp = get_weather_data(i, k, j)
                tf = tf.append(tmp,ignore_index=True)
            except:
                pass

tf.head()
```
